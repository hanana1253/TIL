```python
import requests

url = 'https://apiv2.corona-live.com/updates.json?timestamp=1613657985602'
response = requests.get(url)
response
```

```python
response_dict = response.json()
```

```python
response_dict
```

```python
for item in response_dict:
    cases = item['cases']
    datetime = item['datetime']
    msg = item['src']
    print(f'{datetime}에 {cases}건 발생, 경고메시지: {msg}')
```
